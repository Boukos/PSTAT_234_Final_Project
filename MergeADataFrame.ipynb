```python
import glob
import pandas as pd
import os

# get data file names
path =  os.getcwd()
filenames = glob.glob(path + "/*_data.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)
```

```python
only_close = big_frame[['Name','date','close']]
```

```python
closing_price_5yr = only_close.pivot(index='date', columns='Name', values='close')
```
